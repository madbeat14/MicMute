````python
import sys
import os
# Add src to path if not installed
sys.path.insert(0, os.path.abspath(os.path.join(os.getcwd(), "../src")))

from MicMute.core import AudioController
from MicMute.utils import is_system_light_theme
````

````python
# Initialize Audio Controller (will try to find devices)
audio = AudioController()
print(f"Beep Enabled: {audio.beep_enabled}")
print(f"Current Hotkey: {audio.hotkey_config}")
````

````python
# Check System Theme
is_light = is_system_light_theme()
print(f"System is Light Theme: {is_light}")
````
